# **Read the packages**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import xml.etree.ElementTree as ET
```

# **The location of files**

```python
ann_file_location = 'OBJ_Train_Datasets/Train_Annotations/'
img_file_location = 'OBJ_Train_Datasets/Train_Images/'
out_file_location = 'OBJ_Train_Datasets/Train_Images_STAS_only/'
```

# **Extraction**
We extract the image pieces contains only STAS and store them to the target folder. 

```python
tr_num_list = [f'{i:08}' for i in range(1053)] # We have 1052 samples
extend = 10  # We extend the STAS region for better classification
for tr_num in tr_num_list:
  print(f'Dealing with image {tr_num} now....')
  img_file=img_file_location+f'{tr_num}.jpg'
  ann_file=ann_file_location+f'{tr_num}.xml'
  output_file=f'{tr_num}.jpg'

  # Read the position of the STAS from the annotations file
  tree = ET.parse(ann_file)
  root = tree.getroot()
  STAS_position_dict={} # Iinitialize the dictionary
  count=0               # Initialize the counting number, 
                        # which will represent the number of STAS in each image.

  # Read the image, extract the STAS region and save the image.
  img = cv2.imread(img_file)
  for obj in root.findall('object'):
   xmin=max(int(obj.find('bndbox').find('xmin').text)-extend,0)
   ymin=max(int(obj.find('bndbox').find('ymin').text)-extend,0)
   xmax=min(int(obj.find('bndbox').find('xmax').text)+extend,1716)
   ymax=min(int(obj.find('bndbox').find('ymax').text)+extend,942)
   STAS_img = img[ymin:ymax,xmin:xmax,:]
   cv2.imwrite(out_file_location+tr_num+f'_{count}.jpg',STAS_img,[cv2.IMWRITE_JPEG_QUALITY, 70])
   count+=1
```
