```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import xml.etree.ElementTree as ET
```

```python
t_num_list=[f'{i:08}' for i in range(1053)] # We have 1052 training images.

for t_num in t_num_list:
  print(f'Dealing with image {t_num} now....')
  img_file=f'./OBJ_Train_Datasets/Train_Images/{t_num}.jpg'
  ann_file=f'OBJ_Train_Datasets/Train_Annotations/{t_num}.xml'
  output_file=f'./OBJ_Train_Datasets/Train_Images(labeled)/{t_num}.jpg'

  # Read the position of the STAS from the annotations file
  tree = ET.parse(ann_file)
  root = tree.getroot()
  STAS_position_dict={} # Iinitialize the dictionary
  count=0               # Initialize the counting number, 
                        # which will represent the number of STAS in each image.
  for obj in root.findall('object'):
    xmin=int(obj.find('bndbox').find('xmin').text)
    ymin=int(obj.find('bndbox').find('ymin').text)
    xmax=int(obj.find('bndbox').find('xmax').text)
    ymax=int(obj.find('bndbox').find('ymax').text)
    STAS_position_dict[count]=[(xmin,ymin),(xmax,ymax)]
    count+=1

  # Read the image and position the STAS.
  img = cv2.imread(img_file)
  for obj_num in STAS_position_dict.keys():
    vertex1=STAS_position_dict[obj_num][0]
    vertex2=STAS_position_dict[obj_num][1]
    cv2.rectangle(img,vertex1,vertex2,(0,0,255),3) 
  
  # Save the product
  # cv2.imwrite(output_file,img)

print('finish!')
```
